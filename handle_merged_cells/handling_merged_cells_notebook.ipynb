```python
from openpyxl import load_workbook
import pandas as pd
```

```python
excel_file = r"C:\Users\prash\OneDrive\Desktop\My Work Experience\intelligent-automation-python\handle_merged_cells\merged_cells_example.xlsx"
```

```python
work_book = load_workbook(excel_file)
```

```python
sheet_index = 0
work_sheet = work_book[work_book.sheetnames[sheet_index]]
```

```python
work_sheet
```

```python
processed_data = []
merged_cells = {}

mx_row = work_sheet.max_row
mx_col = work_sheet.max_column

# Create a dictionary of merged cells
for merged_range in work_sheet.merged_cells.ranges:
    for row in range(merged_range.min_row, merged_range.max_row + 1):
        for col in range(merged_range.min_col, merged_range.max_col + 1):
            merged_cells[(row, col)] = merged_range
            
print(f"......Rows : {mx_row} Columns: {mx_col}")
```

```python
merged_cells
```

```python
for row in range(1, mx_row + 1):
    processed_row = [''] * mx_col
    for col in range(1, mx_col + 1):
        if (row, col) in merged_cells:
            merged_range = merged_cells[(row, col)]
            value = work_sheet.cell(merged_range.min_row, merged_range.min_col).value
            
        else:
            value = work_sheet.cell(row=row, column=col).value
        
        if value is not None:
            processed_row[col-1] = str(value)
            
    if any(cell for cell in processed_row):
        if processed_row not in processed_data:
            processed_data.append(processed_row)
            
```

```python
df = pd.DataFrame(processed_data)
```
